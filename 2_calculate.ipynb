```python
import numpy as np
import pandas as pd
import modules.poker_utils as poker
```

```python
subject_map = pd.read_csv('data/raw/subject_map.csv')

# For each subject (this takes a long to run for all 30 – run 1 at a time to test)
for subject_num in range(1, len(subject_map)):
    print(f'STARTING SUBJECT NUMBER {subject_num}')
    villain_name = subject_map.iloc[subject_num - 1, 3]
    hero_name = subject_map.iloc[subject_num - 1, 2]

    game_acts = poker.get_poker_acts(subject_num)
    game_fg = poker.get_fg(game_acts, subject_num)

    #Calculate the poker stats and identify the events in the game
    poker_acts_stats = poker.calc_poker_stats(game_acts, hero_name, villain_name)
    poker_acts_events = poker.name_poker_events(poker_acts_stats, hero_name, villain_name)

    poker_cols = poker_acts_events.columns.values
    FG_cols = game_fg.columns.values

    #Merging the Poker Actions and Facial Gestures
    acts_and_FGs_cols = np.concatenate((FG_cols, poker_cols), axis=None)
    print(acts_and_FGs_cols)
    acts_and_FGs = pd.DataFrame(columns=acts_and_FGs_cols)
    df_index = 0

    #Merging FGs and Poker Acts using the timestamps
    for i in range(0, len(poker_acts_events)):
        start_time_index = poker_acts_events.loc[i, 'RelativeSec']
        if i < len(poker_acts_events) - 1:
            j = i + 1
            end_time_index = poker_acts_events.loc[j, 'RelativeSec']
        else:
            j = len(poker_acts_events) - 1
            end_time_index = poker_acts_events.loc[j, 'RelativeSec']

        temp_FG_df = game_fg[(game_fg['RelativeSec'] > start_time_index) & (
                (game_fg['RelativeSec'] <= end_time_index) |
                (game_fg['RelativeSec'] <= start_time_index))]

        if temp_FG_df.empty:
            d = {'01': [0], '02': [0], '04': [0], '05': [0], '06': [0], '07': [0], '09': [0], '10': [0], '12': [0],
                 '14': [0], '15': [0], '17': [0], '20': [0], '23': [0], '25': [0], '26': [0], '45': [0],
                 'Inflection': [0], 'Onset Length': [0], 'Onset Unity': [0], 'Offset Length': [0], 'Offset Unity': [0],
                 'SumAmp': [0], 'TrueInflection': [0], 'RelativeTime': [0], 'RelativeSec': [0]}
            temp_FG_df = pd.DataFrame(data=d)

        for k in range(0, len(temp_FG_df)):
            df_index += 1

            new_row = np.concatenate((temp_FG_df.iloc[k, :].values, poker_acts_events.iloc[i, :].values), axis=None)

            acts_and_FGs.loc[df_index] = new_row
    acts_and_FGs.to_csv(f'data/processed/acts_and_FGs_NEW(PFLP,CRF)_{subject_num}.csv')
```
